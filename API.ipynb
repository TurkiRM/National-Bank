# Step I: Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import requests
import pandas as pd
import json
import numpy as np
response_API = requests.get('https://search.worldbank.org/api/v2/projects')
#print(response_API.status_code)
```

# Step II: Assigning the text content of the API response to the variable 'data'


```python
data = response_API.text
```

```python
data
```

# Step III: Parsing the JSON-formatted text stored in the variable 'data' into a Python dictionary or object


```python
jdata = json.loads(data)
```

# Step IV:  Extracting project-related data from the parsed JSON data


```python
projects_data = jdata['projects']
```

# Step V: Converting the project-related data into a DataFrame and transposing it


```python
df = pd.DataFrame(projects_data).transpose()
```

### Set the index

```python
df['id'].is_unique
```

```python
df.index = df.index.astype(str)
```

```python
# df.set_index(df.columns[3], inplace=True)
df.set_index('id', inplace=True)
```
